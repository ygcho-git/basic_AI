# 경사하강법

image.png

```python
# 절편을 79.0이라 하고, a값에 따른 MSE값을 그래프로 표현

import numpy as np
import matplotlib.pyplot as plt

X = np.array([2, 4, 6, 8])
Y = np.array([81, 93, 91, 97])

A = np.linspace(-2, 6.5)

y = []

for a in A : 
    Y_pred = a * X + 79.0
    error = Y_pred - Y
    mse = np.mean(error ** 2)
    y.append(mse)


plt.scatter(A, y)
plt.xlabel("a")
plt.ylabel("MSE")
plt.show()
```

```python
# 인간 경사하강법
import numpy as np
 
X = np.array([2, 4, 6, 8])
Y = np.array([81, 93, 91, 97])
 
a = 0
b = 79.0
n = 4
lr = 0.03

Y_pred = a * X + b

error = Y_pred - Y
MSE = np.mean(error**2)

a_grad = (2/n) * np.sum(error * X) 

print(a, a_grad, MSE)
```

경사하강법 : 함수 값이 낮아지는 방향으로 값을 변형시켜 최종적으로 최소(함수)값을 갖도록 하는 값을 찾는 법

image.png

그래프 위에 랜덤하게 파라미터가 정해지면, 해당 파라미터의 위치로 부터 최소값을 찾아 탐색하는 기준으로 경사도를 구해야 함. 

빨간점의 위치일 때 : 경사의 기울기가 음수(-)이므로, a의 값을 증가

파란점의 위치일 때 : 경사의 기울기가 양수(+)이므로, a의 값을 감소

경사의 기울기를 알기 위해선, 미분이 필요하다.

```python
import numpy as np
 
X = np.array([2, 4, 6, 8])
Y = np.array([81, 93, 91, 97])
 
a = 0
b = 79.0
n = 4
lr = 0.03

Y_pred = a * X + b

error = Y_pred - Y
MSE = np.mean(error**2)

a_grad = (2/n) * np.sum(error * X) 

print(a, a_grad, MSE)
```
