## 회귀 : 독립변수($X$)들과 종속변수($Y$)들의 관계를 나타내는 함수(모델)을 찾는 것.

> 단순회귀 : $x$가 1개(예, 공부시간과 성적) $\rightarrow$ $y = ax + b$  
> 다중회귀 : $x$가 2개 이상(예, (공부시간, 과외시간)과 성적) $\rightarrow$ $y = a_1x_1 + a_2x_2 + b$

## 지금부터는 $ax+b$가 아니라 $wx+b$로...
> $y=a_1x_1 + a_2x_2 + a_3x_3 + ... + a_nx_n + b$ $→$  
> $y=w_1x_1 + w_2x_2 + w_3x_3 + ... + w_nx_n + b$

> 가중치(weight) w
>> 각 입력 신호가 결과 출력에 미치는 **중요도**를 조절하는 매개변수  

> 편향(bias) b
>> 뉴런의 **활성화를 조절**하는 매개변수 - 값의 보정  
>> 뉴런의 활성화 조건을 설정하는 매개변수

## 단순 선형 회귀

```python
# y = wx+b = 2.3x+79.0
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로우, 케라스 함수 import
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model

X = np.array([2, 4, 6, 8])
Y = np.array([81, 93, 91, 97])

# 모델의 구조 설계
model = Sequential()
#model.add(Dense(1, input_dim=?, activation='linear'))
model.compile(optimizer='sgd', loss='mse')    
# optimizer : 오차를 어떻게 보정하여 최적화된 w와 b를 구할 것인가?
# loss : 손실함수

#오차를 최소화하는 과정을 2000번 반복
model.fit(X, Y, epochs=2000)
```

## 다중 선형 회귀

```python
# y = w1*x1 + w2*x2 + b = 1.51 * x1 + 2.29 * x2 + 77.72
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로우, 케라스 함수 import
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model

X = np.array([[2,0], [4, 4], [6, 2], [8, 3]])
Y = np.array([81, 93, 91, 97])

# 모델의 구조 설계
model = Sequential()
#model.add(Dense(1, input_dim=?, activation='linear'))
model.compile(optimizer='sgd', loss='mse')    

model.fit(X, Y, epochs=20)
```
