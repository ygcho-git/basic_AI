```python
# 필요한 모듈 불러오기
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
import pandas as pd

# 데이터 다운받기
(train_x, train_y), (test_x, test_y) = mnist.load_data()

# 데이터 전처리
# train_x = train_x.reshape(60000, 784)   # 제거  --->   # Flattern() 레이어를 추가하여 해결
# train_y = pd.get_dummies(train_y)       # 제거  --->   # label이 int 타입일 경우, one-hot encoding 하지 않고,
                                          # 손실함수를 sparse_categorical_crossentropy를 사용해도 됨. 성능상의 차이 없는 것으로 알려짐.
train_x = train_x / 255.0                 # 정규화 - 값의 범위를 조정 --> 학습 속도가 향상

# 모델 설계
model = Sequential()
model.add(Input(shape=(28, 28)))
model.add(Flatten())                        # Flatten 레이어에서 (60000, 25, 25) --> (60000,784)
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(train_x, train_y, epochs=10)
```
